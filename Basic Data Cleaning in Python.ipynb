```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.columns
```

```python
a = sns.countplot(x='Gender',data=df)
for i in a.containers:
    a.bar_label(i)
```

```python
b = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=b)
```

```python
df.columns
```

```python
c=sns.countplot(x='Age Group',hue='Gender',data=df)
c
for i in c.containers:
    c.bar_label(i)
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
df.columns
```

```python
state_order = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='State',y='Orders',data=state_order)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
state_amt = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='State',y='Amount',data=state_amt)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
df.columns
```

```python
count=sns.countplot(x='Marital_Status',data=df)
count
sns.set(rc={'figure.figsize':(6,5)})
for i in count.containers:
    count.bar_label(i)
```

```python
m_amt = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=m_amt)
```
